# Guess Game

```python

from random import randint
random_number=randint(1,10) # 
#random_number
guess = None
while True:
    guess = input("Pick a number from 1 to 10: ")
    guess=int(guess)
    if guess < random_number:
        print("TOO LOW!")
    elif guess > random_number:
        print("TOO HIGHT")
    else:
        print("You GOT IT!")
        play_again = input("Do you want to play again? (y/n) ")
        if play_again =='y':
            random_number=randint(1,10)
            guess=None
        else:
            print("Thank you")
            break
```

```python
# Rock,Paper , Scissor
from random import randint
player_wins =0
computer_wins =0
play_number=2
while player_wins <play_number and computer_wins<play_number:
    print(f"Player Score: {player_wins}, Computer Score: {computer_wins}")
    print("...rock...")
    print("...paper...")
    print("...scissors...")
    player1=input("Please input your choice: ").lower()
    if player1=="quit" or player1=="q":
        break
    rand_num=randint(0,2)
    #rand_num
    if player1:
        if rand_num ==0:
            computer="rock"
        elif rand_num==1:
            computer="paper"
        else:
            computer="scissor"
        print(f"The computer plays {computer}")

        if player1==computer:
            print("you are tied")
        elif player1=="rock":
            if computer=="scissor":
                print("player1 wins")
                player_wins+=1
            else:
                print("computer wins")
                computer_wins+=1
        elif player1=="paper":
            if computer=="scissor":
                print("computer wins")
                computer_wins+=1
            else:
                print("player1 wins")
                player_wins+=1
        elif player1=="scissor":
            if computer=="rock":
                print("computer wins")
                computer_wins+=1
            else:
                print("player1 wins")
                player_wins+=1
        
    else:
        print("Please input your choice!")
        
if player_wins>computer_wins:
    print("YOU WIN!!!")
elif player_wins==computer_wins:
    print("IT'S A TIE")
else:
    print(":( What a pity!!!")
print(f"Player Score: {player_wins}, Computer Score: {computer_wins}")
```
